```python
# load numpy
import numpy as np
# load pandas 
import pandas as pd
csv_file ="C:/Users/Maitreyi/Downloads/Upgrad/Assignment/States.csv"
# read cvs with pandas read_csv

df = pd.read_csv(csv_file)
df = df.drop('West Bengal1', axis=1)
df
```

```python
## drop item Description
MDF = df.drop(df.columns[[0]], axis=1) 
## drop rows
MDF =MDF.drop([0,1,2,3,4,5,6,10] )
MDF
```

```python
MDF.set_index('Duration' , inplace = True)
MDF
```

```python
mait_growth_df =MDF.transpose()
mait_growth_df['Avg Growth Rate'] = mait_growth_df.mean(axis = 1)
mait_growth_df
```

```python
mait_growth_df.index.name ='state'
mait_growth_df = mait_growth_df.sort_values('Avg Growth Rate' , ascending = False)
mait_growth_df
mait_growth_df_fast = mait_growth_df[(mait_growth_df['Avg Growth Rate'] > 12) ]
mait_growth_df_fast
```

```python
mait_growth_df_low = mait_growth_df[(mait_growth_df['Avg Growth Rate'] < 12) ]
mait_growth_df_low
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns


sns.set(style = "white")
## fastest growing state
plt.figure(figsize =(20,15))
sns.barplot(x =  'Avg Growth Rate' , y = mait_growth_df_fast.index ,data = mait_growth_df_fast   )

plt.ylabel('States' ,fontsize ='20 ')
plt.xlabel('Avg Growth Rate > 12' ,fontsize ='20')
plt.title('States Growing Fast' ,fontsize ='30')
plt.xlim(0,20)

## Slowest growing state

plt.figure(figsize =(20,15))
sns.barplot(x =  'Avg Growth Rate' , y = mait_growth_df_low.index ,data = mait_growth_df_low )

plt.ylabel('States' ,fontsize ='20' )
plt.xlabel('Avg Growth Rate < 12' ,fontsize ='20' )
plt.title('States Growing Low' ,fontsize ='30')
plt.xlim(0,20)

plt.show()
```

```python
##Plot the total GDP of the states for the year 2015-16
df = df.drop(df.columns[[0]], axis=1) 
GDF =df.drop([0,1,2,3,5,6,7,8,9,10] )
GDF = GDF.transpose()
GDF = GDF.dropna()
GDF.index.name = 'State'
GDF.rename(columns={4 : 'GDP'} ,inplace =True)
GDF = GDF.sort_values('GDP' ,ascending =False)
GDF.drop(GDF.head(1).index, inplace=True)
GDF
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns


sns.set(style = "white")
## fastest growing state
plt.figure(figsize =(20,15))
sns.barplot(x =  'GDP' , y = GDF.index ,data = GDF   )

plt.ylabel('States' ,fontsize ='20 ')
plt.xlabel('Gdp over 2015 &16' )
plt.title('States Gdp over 2015 & 16' )
plt.show()
```

```python
##Identify the top-5 states based on total GDP
GDF_Top = GDF.head(5)
GDF_Top
```
